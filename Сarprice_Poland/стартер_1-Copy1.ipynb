–≠—Ç–æ —Ä–µ–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ, —Å–æ–±—Ä–∞–Ω–Ω—ã–µ –≤ –ü–æ–ª—å—à–µ.

## –û—á–µ–Ω—å –ø—Ä–æ—Å—Ç–∞—è –º–æ–¥–µ–ª—å
–î–∞–≤–∞–π —Å—Ä–∞–∑—É —Å–¥–µ–ª–∞–µ–º "—Ä—É—á–Ω—É—é" –≤–µ—Ä—Å–∏—é –º–æ–¥–µ–ª–∏, —á—Ç–æ–±—ã –±—ã–ª–æ –±–æ–ª–µ–µ –ø–æ–Ω—è—Ç–Ω–æ. –î–ª—è —ç—Ç–æ–≥–æ –Ω–∞—Ö–æ–¥–∏–º —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏ –≤—Å–µ–≥–¥–∞ –≤–æ–∑–≤—Ä–∞—â–∞–µ–º –µ–≥–æ –¥–ª—è –Ω–∞—à–µ–≥–æ "–ø—Ä–æ–≥–Ω–æ–∑–∞". –í–æ—Ç —Ç–∞–∫ –ø—Ä–æ—Å—Ç–æ.

–ù–∞—à –ø—Ä–æ–≥–Ω–æ–∑ –∑–∞–ø–∏—Å—ã–≤–∞–µ–º –≤ —Å—Ç–æ–ª–±—Ü–µ `price_pred`.

```python
price_mean = df["price_value"].mean()
df["price_pred"] = price_mean
price_mean
```

```python
df.sample(5)
```

–¢–µ–ø–µ—Ä—å –¥–∞–≤–∞–π —Å—á–∏—Ç–∞—Ç—å –æ—à–∏–±–∫—É. –£ –Ω–∞—Å –µ—Å—Ç—å –ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –æ—Ç–≤–µ—Ç (–≤ `price_value`) –∏ —Ç–æ, —á—Ç–æ –º—ã —Å–ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–ª–∏ (–≤ `price_pred`). –¢–µ–ø–µ—Ä—å, –≤—ã—á–∏—Ç–∞—è –æ–¥–Ω–æ –∏–∑ –¥—Ä—É–≥–æ–≥–æ -  –ø–æ–ª—É—á–∞–µ–º —Ä–∞–∑–Ω–∏—Ü—É (–æ—à–∏–±–∫—É).

–ó–∞—Ç–µ–º "–æ–±—Ä–µ–∑–∞–µ–º –∑–Ω–∞–∫" –∏ –Ω–∞—Ö–æ–¥–∏–º —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –æ—à–∏–±–∫–∏.

```python
df["price_diff"] = df["price_pred"] - df["price_value"]

df["price_diff"].abs().mean()
```

–ü–æ–ª—É—á–∞–µ—Ç—Å—è, —á—Ç–æ –æ—à–∏–±–∫–∞ - —á—É—Ç—å –±–æ–ª—å—à–µ **40 522 PLN (–ø–æ–ª—å—Å–∫–∏—Ö –∑–ª–æ—Ç—ã—Ö)**. –≠—Ç–æ –∑–Ω–∞—á–∏—Ç, —á—Ç–æ –Ω–∞—à–∞ –º–æ–¥–µ–ª—å –≤—Å–µ –µ—â–µ —Ä–∞–±–æ—Ç–∞–µ—Ç –Ω–µ –æ—á–µ–Ω—å —Ö–æ—Ä–æ—à–æ. –ù–æ —Å –¥—Ä—É–≥–æ–π —Å—Ç–æ—Ä–æ–Ω—ã, –ø–æ–ª–µ–∑–Ω–æ –¥–µ–ª–∞—Ç—å —Ç–∞–∫–∏–µ –ø—Ä–æ—Å—Ç—ã–µ –º–æ–¥–µ–ª–∏, —á—Ç–æ–±—ã –±—ã–ª–æ —Å —á–µ–º —Å—Ä–∞–≤–Ω–∏–≤–∞—Ç—å —Å–ª–µ–¥—É—é—â–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç.

–ù–∞ —É–¥–∏–≤–ª–µ–Ω–∏–µ, –≤ —Ä–µ–∞–ª—å–Ω–æ–π –∂–∏–∑–Ω–∏ –≤–æ—Ç —Ç–∞–∫–∏–µ –ø—Ä–æ—Å—Ç—ã–µ –º–æ–¥–µ–ª–∏ –º–æ–≥—É—Ç –Ω–µ —Å–∏–ª—å–Ω–æ –æ—Ç–ª–∏—á–∞—Ç—å—Å—è –æ—Ç –æ—á–µ–Ω—å "—Å–ª–æ–∂–Ω—ã—Ö" –º–æ–¥–µ–ª–µ–π. –ü–æ—ç—Ç–æ–º—É –ø–æ–ª–µ–∑–Ω–æ –∑–Ω–∞—Ç—å, –∫–∞–∫–æ–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –º–æ–∂–µ–º –¥–æ—Å—Ç–∏—á—å "–Ω–∏—á–µ–≥–æ –Ω–µ –¥–µ–ª–∞—è" (–≤–æ–∑–≤—Ä–∞—â–∞—è –ø—Ä–æ—Å—Ç–æ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ).

–ö—Å—Ç–∞—Ç–∏, —á—Ç–æ –∫–∞—Å–∞–µ—Ç—Å—è –º–µ—Ç—Ä–∏–∫–∏ - —Ç–æ, —á—Ç–æ –º—ã —Å–¥–µ–ª–∞–ª–∏ –≤—ã—à–µ - –∏–º–µ–µ—Ç –Ω–∞–∑–≤–∞–Ω–∏–µ - `mean_absolute_error`.

```python
mean_absolute_error(df["price_value"], df["price_pred"])
```

–î–∞–ª–µ–µ –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —É–∂–µ –≥–æ—Ç–æ–≤—É—é —Ñ—É–Ω–∫—Ü–∏—é `mean_absolute_error`, –Ω–æ –Ω–∞–¥–µ—é—Å—å, —É–∂–µ —Å—Ç–∞–ª–æ –ø–æ–Ω—è—Ç–Ω–æ - –∫–∞–∫ –æ–Ω–æ —Å—á–∏—Ç–∞–µ—Ç.

–ê —á—Ç–æ –µ—Å–ª–∏ –≤–æ–∑–≤—Ä–∞—â–∞—Ç—å –Ω–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, –∞ –º–µ–¥–∏–∞–Ω—É (–≤ —à–ø–∞—Ä–≥–∞–ª–∫–µ –ø–æ [pandas](—à–ø–∞—Ä–≥–∞–ª–∫–∞_pandas.ipynb) –Ω–∞–π–¥–µ—à—å —Ä–∞–∑–Ω–∏—Ü—É –º–µ–∂–¥—É —ç—Ç–∏–º–∏ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è–º–∏).

```python
df["price_pred"] = df["price_value"].median()
mean_absolute_error(df["price_value"], df["price_pred"])
```

–ö–∞—á–µ—Å—Ç–≤–æ –º–æ–¥–µ–ª–∏ —É–ª—É—á—à–∏–ª–æ—Å—å. –ö—Å—Ç–∞—Ç–∏, —Ç–æ, —á—Ç–æ –º—ã —Å–µ–π—á–∞—Å –¥–µ–ª–∞–ª–∏ –≤—Ä—É—á–Ω—É—é, –º–æ–∂–Ω–æ –¥–æ—Å—Ç–∏—á—å –ø—Ä–∏ –ø–æ–º–æ—â–∏ –≥–æ—Ç–æ–≤—ã—Ö –º–æ–¥–µ–ª–µ–π. –ë–ª–∞–≥–æ–¥–∞—Ä—è —ç—Ç–æ–º—É, —Ç–æ–∂–µ —Å–æ—Ö—Ä–∞–Ω–∏–º "—Å—Ç–∞–Ω–¥–∞—Ä—Ç" scikit-learn –∏ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –ª–µ–≥–∫–æ –ø–æ–¥–º–µ–Ω—è—Ç—å –º–æ–¥–µ–ª–∏.

## –°–æ–∑–¥–∞–µ–º `X` –∏ `y`.
`Dummy` –º–æ–¥–µ–ª—å, —á—Ç–æ –≤ –ø–µ—Ä–µ–≤–æ–¥–µ —Å –∞–Ω–≥–ª–∏–π—Å–∫–æ–≥–æ –æ–∑–Ω–∞—á–∞–µ—Ç "–≥–ª—É–ø–µ–Ω—å–∫–∞—è/–¥—É—Ä–æ—á–∫–æ–≤–∞—Ç–∞—è" –º–æ–¥–µ–ª—å.

```python
feats = ["id"]
X = df[feats]
y = df["price_value"]
```

## –¢—Ä–µ–Ω–∏—Ä—É–µ–º –º–æ–¥–µ–ª—å –∏ –∏–∑–º–µ—Ä—è–µ–º –∫–∞—á–µ—Å—Ç–≤–æ
–î–ª—è –Ω–∞—á–∞–ª–∞ –≤—Å—Ç–∞–≤–∏–º `strategy="mean"`.

```python
model = DummyRegressor(strategy="mean")
model.fit(X, y)
y_pred = model.predict(X)

mean_absolute_error(y, y_pred)
```

–í–∏–¥–∏–º, —á—Ç–æ –∫–∞—á–µ—Å—Ç–≤–æ –æ—Å—Ç–∞–ª–æ—Å—å —Ç–æ –∂–µ —Å–∞–º–æ–µ (–∫–∞–∫ –∏ –≤ —Ä—É—á–Ω–æ–π –≤–µ—Ä—Å–∏–∏ —Å–æ —Å—Ä–µ–¥–Ω–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ–º). –¢–µ–ø–µ—Ä—å –ø–æ–Ω–∏–º–∞–µ—à—å, –∫–∞–∫ —ç—Ç–æ —Ä–∞–±–æ—Ç–∞–µ—Ç –≤–Ω—É—Ç—Ä–∏.

–ü—Ä–æ–≤–µ—Ä–∏–º –µ—â–µ –¥–ª—è –º–µ–¥–∏–∞–Ω—ã.

```python
model = DummyRegressor(strategy="median")
model.fit(X, y)
y_pred = model.predict(X)

mean_absolute_error(y, y_pred)
```

–í—Ä–µ–º—è —Å–¥–µ–ª–∞—Ç—å –ø—Ä–æ–≥–Ω–æ–∑ –∏ —Å–¥–µ–ª–∞—Ç—å –ø–µ—Ä–≤—ã–π "submit" –≤ Kaggle, —á—Ç–æ–±—ã –ø–æ–Ω—è—Ç—å, –∫–∞–∫ —ç—Ç–æ —Ä–∞–±–æ—Ç–∞–µ—Ç.

```python
df_test = pd.read_hdf("../input/df.test.h5")

X_test = df_test[feats]
df_test["price_value"] = model.predict(X_test)

df_test[["id", "price_value"]]
```
