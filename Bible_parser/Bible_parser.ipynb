Старый код

```python
import requests
from lxml import html
url = 'https://cgamos.ru/images/archive/01-0203-0768-000152/'

page = requests.get(url)
tree = html.fromstring(page.content)

a = [x.attrib['src'] for x in tree.cssselect('img') if 'JPG' in x.attrib['src']]

for pic_link in a:
    with open('test/'+pic_link.split('/')[-1], 'wb') as f:
        f.write(requests.get(pic_link).content)
```

```python
import requests 
import pandas as pd
from bs4 import BeautifulSoup # Библиотека BeautifulSoup используется для не динамического парсинга

url = "https://cgamos.ru/metric-books/203/203-768/203_768_152/?sphrase_id=107956"
result = requests.get(url)
```

```python
if result.status_code == 200:
    soup = BeautifulSoup(result.text, "html.parser")
```

```python
images = soup.find_all('img') # скачиваем все атрибуты картинок на этом сайте. 
images
```

```python
type(images) # специальный формат библиотеки, откуда можно  сразу выташить все ссылки (тег src=""/)
# но так как ссылка каждой фотографии не активна пока не перейдут на ее страницу (все src=""/ как видно, пустые), 
# то мы будем вытаскивать ссылки просто из текстового фаила
```

```python
lst = [] # создаем список с текстом, откуда потом будет вытаскивать ссылки
# на картинки, удаляя ненужные (банеры сверху и снизу сайта)
lst = images[0:len(images)]
del lst[:2]
del lst[-1]
lst
```

```python
len(lst)
```

```python
lst[236]
```

```python
image_url = [] # далее мы из текса берем названия картинок с помощью уникальных резделителей
for i in range(len(lst)):
    names = str(lst[i]).split('/images/archive/01-0203-0768-000152/')[1]
    name = names.split('" src=""/>')[0]
    image_url.append(name)
image_url[0] = image_url[0].split('" src="')[0] # 1ый элемент отличается, так как он открыт, для него доп. обработка
```

```python
image_url 
```

```python
image_url_full = [] # прибавляем название к ссылке сайта, получая тем самым полноценную ссылку на картинку
for i in range(len(image_url)):
    image_url_full.append('https://cgamos.ru/images/archive/01-0203-0768-000152/' + image_url[i])
```

```python
image_names = [] # отделяем .jpg от названий
for i in range(len(image_url)):
    names = str(image_url[i]).split('.')[0]
    image_names.append(names)
```

```python
import urllib # сохраняем все картинки в папку где лежит этот ноутбук. картинок много, так что следует заранее позаботиться о своболном месте
for i in range(len(image_url_full)):
    urllib.request.urlretrieve(image_url_full[i], image_names[i]+".jpg")
```
