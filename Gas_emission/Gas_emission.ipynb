```python
import pandas as pd
import numpy as np
import os, json
from pandas.io.json import json_normalize
import statistics as st
```

```python
df = pd.read_json('4day.json')
# df
```

```python
dot_list = [] # Список названий точек измерений
for i in df['datasets']:
    for new_s, new_val in i.items():
        dot_list.append(new_s)
```

```python
j = 0 # Проход по каждой точке
df_result = pd.DataFrame()
for item in df["datasets"]:
    dict_value = item[dot_list[j]]['reps']["REP_1"]['data']
    columns_names = dict_value.keys()
    
    df_value = pd.DataFrame(dict_value.items())
    df_value = df_value.T
    df_value.columns = df_value.iloc[0]
    df_value = df_value[1:]
    df_value = df_value.set_index(['timestamp']).apply(pd.Series.explode).reset_index()
    df_value.drop(columns=['timestamp'], inplace = True)
    
    df_result = df_result.append(df_value.mean().to_frame().T)

    j += 1
df_result['Dot'] = dot_list
```

```python
j = 0 # Добавление координат точек
latitude_list = []
longitude_list = []
for item in df["datasets"]:
    latitude = item[dot_list[j]]['reps']["REP_1"]['header']['latitude']
    latitude_list.append(latitude)
    
    longitude = item[dot_list[j]]['reps']["REP_1"]['header']['longitude']
    longitude_list.append(longitude)

    j += 1
df_result['latitude'] = latitude_list
df_result['longitude'] = longitude_list
```

```python
df_total = pd.DataFrame()
for file in json_files:
        df = pd.read_json(file)
        dot_list = [] # Список названий точек измерений
        for i in df['datasets']:
            for new_s, new_val in i.items():
                dot_list.append(new_s)

#         print(file)
#         print(dot_list)
        j = 0
        df_result = pd.DataFrame()
        dot_names = []
        latitude_list = []
        longitude_list = []
        for item in df["datasets"]:
            try:
                dict_value = item[dot_list[j]]['reps']["REP_1"]['data']
                columns_names = dict_value.keys()

                df_value = pd.DataFrame(dict_value.items()) #Экспорт основных параметров
                df_value = df_value.T
                df_value.columns = df_value.iloc[0]
                df_value = df_value[1:]
                df_value = df_value.set_index(['timestamp']).apply(pd.Series.explode).reset_index()
                df_value.drop(columns=['timestamp'], inplace = True)

                df_result = df_result.append(df_value.mean().to_frame().T)
                dot_names.append(dot_list[j])

                
                latitude = item[dot_list[j]]['reps']["REP_1"]['header']['latitude'] # Добавление широты/долготы
                latitude_list.append(latitude)
                longitude = item[dot_list[j]]['reps']["REP_1"]['header']['longitude']
                longitude_list.append(longitude)
#                 print(dot_names)
#                 print(dict_value.items())
#                 print(j)
            except KeyError:
                pass
            j += 1
        df_result.insert(loc=0, column='Day', value=file)
        df_result.insert(loc=1, column='Dot', value=dot_names)
        df_result.insert(loc=2, column='latitude', value=latitude_list)
        df_result.insert(loc=3, column='longitude', value=longitude_list)
        df_total = df_total.append(df_result, ignore_index=True)
df_total.dropna(subset=['ch4', 'ch4'], inplace = True)
```
