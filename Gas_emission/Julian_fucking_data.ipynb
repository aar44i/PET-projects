```python
import pandas as pd
import numpy as np
from numpy import int32
from datetime import datetime, timedelta
```

```python
data = pd.read_excel('julian_data.xlsx')
```

```python
df = data.copy()
```

```python
df.rename(columns={"Unnamed: 44": "datetime"}, inplace = True)
```

```python
df.info()
```

```python
# df['int_date'] = int(df['Julian Day'])
df['int_date'] = df['Julian Day'].astype(int)
```

```python
df.sample(5)
```

```python
df.info()
```

```python
df.head()
```

```python
for i in range(len(df)):
    if df['int_date'].loc[i+1] != df['int_date'].loc[i]:
        df['datetime'].loc[i+1] = df['datetime'].loc[i] + timedelta(days = int(df['int_date'].loc[i+1] - df['int_date'].loc[i]))
    else:
        df['datetime'].loc[i+1] = df['datetime'].loc[i]


    
```

```python
# df.to_excel('all_with_data.xlsx')
```

```python
df_group = df.groupby([df['datetime'].dt.date]).mean()
```

```python
df_group.to_excel('all_with_data_gro.xlsx')
```
