```python
import pandas as pd
from pandas_profiling import ProfileReport
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default='notebook'
```

```python
df = pd.read_csv('data.csv', sep = ";", names = ['date', 'value'], skiprows=1)
df
```

```python
df.info() 
```

```python
## создаём фигуру при помощи plotly
fig = go.Figure()
## добавляем на неё график
fig.add_trace(go.Scatter(
    x=df.date, y=df.value,
    name='EURUSD',line_color='ForestGreen'
))
# # подписываем титул и оси
fig.update_layout(
    title=go.layout.Title(text="Расчетные счета клиентов Сбербанка", x=0.5),
    xaxis=go.layout.XAxis(title=go.layout.xaxis.Title(text="Дата")),
    yaxis=go.layout.YAxis(title=go.layout.yaxis.Title(text="Руб."))
    )
```
