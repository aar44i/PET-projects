```python
df_mean_model = pd.read_excel(open('тренды_факт_модель_1977_2018.xlsx', 'rb'), sheet_name='сред.год.модель')
df_mean_fact = pd.read_excel(open('тренды_факт_модель_1977_2018.xlsx', 'rb'), sheet_name='сред.год.факт')
df_max_model = pd.read_excel(open('тренды_факт_модель_1977_2018.xlsx', 'rb'), sheet_name='макс.год.модель')
df_max_fact = pd.read_excel(open('тренды_факт_модель_1977_2018.xlsx', 'rb'), sheet_name='макс.год.факт')
```

```python
columns = list(df_mean_model.columns)
columns.remove('Год')
columns
```

```python
for i in columns:
    fig, (ax1, ax2) = plt.subplots(figsize=(17, 5),ncols=2,sharex=False, sharey=False)
    line1, = ax1.plot(df_mean_fact['Год'],df_mean_fact[i], label='Фактические данные')
    line2, = ax1.plot(df_mean_model['Год'],df_mean_model[i], label='ECOMAG', color = 'red')
    ax1.legend(handles=[line1, line2])
    ax1.set_title('а)')
    ax1.set_ylabel('Расход воды [м3/с]')


    line1, = ax2.plot(df_max_fact['Год'],df_max_fact[i], label='Фактические данные')
    line2, = ax2.plot(df_max_model['Год'],df_max_model[i], label='ECOMAG', color = 'red')
    ax2.legend(handles=[line1, line2])
    ax2.set_title('б)')
    ax2.set_ylabel('Расход воды [м3/с]')
    fig.suptitle(i, fontsize=16)
    plt.savefig(i, dpi=800)
```
