```python
import numpy as np
from scipy.stats import variation
import pandas as pd 
import seaborn as sns
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('ETR Qyear.xlsx')
df = df.fillna(0)
df.head()
```

```python
df_trans = df.T
df_trans.columns = df_trans.iloc[0]
df_trans = df_trans.iloc[1:]
df_trans['mean'] = df.mean()
df_trans.reset_index(inplace = True)
df_trans.rename(columns={'index': 'gauge'}, inplace=True)
df_trans
```

```python
df = df_trans # просто потому что мне лень все перемннные переименовывать
gauges = df.gauge 
```

```python
gauges_names = list(df.columns.values) # тут теперь сохраняются года, но мне лень переименовывать
gauges_names.pop()
del gauges_names[0]
for n, i in enumerate(gauges_names):
    if i == '83323.1':
        gauges_names[n] = 83323
gauges_names
```

```python
%%time
df_modul_coef = pd.DataFrame()
c = 0
for i in(gauges_names):
#     print(i, c)
    df_modul_coef[i] = df[i] / df['mean'] - 1
    c = c + 1
```

```python
df_modul_coef
```

```python
df_modul_coef = df_modul_coef.replace(-1, np.nan)
df_modul_coef['gauges'] = gauges
cols = list(df_modul_coef.columns)
cols = [cols[-1]] + cols[:-1] 
df_modul_coef = df_modul_coef[cols]
df_modul_coef
```
